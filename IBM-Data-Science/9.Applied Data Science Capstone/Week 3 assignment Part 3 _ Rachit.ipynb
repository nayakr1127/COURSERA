Showing postal codes on Maps

```python
torontomap = folium.Map(location=[43.65, -79.4], zoom_start=12)

X = torontocodes['Latitude']
Y = torontocodes['Longitude']
Z = np.stack((X, Y), axis=1)

kmeans = KMeans(n_clusters=4, random_state=0).fit(Z)
clusters = kmeans.labels_
colors = ['red', 'green', 'blue', 'yellow']
torontocodes['Cluster'] = clusters
for latitude, longitude, borough, cluster in zip(torontocodes['Latitude'], torontocodes['Longitude'], torontocodes['Borough'], torontocodes['Cluster']):
    label = folium.Popup(borough, parse_html=True)
    folium.CircleMarker([latitude, longitude],
                        radius=5,
                        popup=label,
                        color='black',
                        fill=True,
                        fill_color=colors[cluster],
                        fill_opacity=0.7).add_to(torontomap)  
torontomap
```
